```python
# Task 1

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:
    # 1. Clean the name, roll, and marks
    clean_name = student["name"].strip().title()
    clean_roll = int(student["roll"])
    clean_marks = [int(mark) for mark in student["marks_str"].split(", ")]

    # 2. Validate the name (check if every word contains only alphabetic characters)
    is_valid = all(word.isalpha() for word in clean_name.split())
    validation_status = "✓ Valid name" if is_valid else "✗ Invalid name"

    # Print the validation status next to the student's name
    print(f"{clean_name} {validation_status}")

    # Store the cleaned dictionary for later use
    cleaned_students.append({
        "name": clean_name,
        "roll": clean_roll,
        "marks": clean_marks
    })

    # 3. Print the formatted profile card
    print("================================")
    print(f"Student : {clean_name}")
    print(f"Roll No : {clean_roll}")
    print(f"Marks   : {clean_marks}")
    print("================================\n")

# 4. Print the name in ALL CAPS and lowercase for the student with roll number 103
for student in cleaned_students:
    if student["roll"] == 103:
        print(student["name"].upper())
        print(student["name"].lower())
```

```python
# Task 2

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

# 1. Print subjects, marks, and grades using a for loop
print(f"--- Grades for {student_name} ---")
for subj, mark in zip(subjects, marks):
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"
    print(f"{subj}: {mark} - Grade {grade}")

# 2. Calculate and print summary statistics
total_marks = sum(marks)
average_marks = total_marks / len(marks)

# Find the highest and lowest marks, then use their index to find the corresponding subject
highest_mark = max(marks)
highest_index = marks.index(highest_mark)
highest_subject = subjects[highest_index]

lowest_mark = min(marks)
lowest_index = marks.index(lowest_mark)
lowest_subject = subjects[lowest_index]

print(f"\nTotal marks: {total_marks}")
print(f"Average marks: {average_marks:.2f}")
print(f"Highest scoring subject: {highest_subject} ({highest_mark})")
print(f"Lowest scoring subject: {lowest_subject} ({lowest_mark})\n")

# 3. Simulate marks-entry system using a while loop
print("--- Add New Subjects ---")
new_subjects_count = 0

while True:
    new_subj = input("Enter subject name (or type 'done' to stop): ").strip()

    # Check for the exit condition
    if new_subj.lower() == 'done':
        break

    new_mark_input = input(f"Enter marks for {new_subj} (0-100): ").strip()

    # Validate input: Check if it's a valid integer
    try:
        new_mark = int(new_mark_input)
    except ValueError:
        print("Warning: Invalid input. Please enter a numeric value.")
        continue # Skips the rest of the loop and starts the next iteration

    # Validate input: Check if it's within the 0-100 range
    if new_mark < 0 or new_mark > 100:
        print("Warning: Marks must be between 0 and 100.")
        continue

    # If validations pass, append the new data
    subjects.append(new_subj)
    marks.append(new_mark)
    new_subjects_count += 1
    print(f"✓ Added {new_subj} with {new_mark} marks.")

# 4. Print the final summary after the loop ends
print("\n--- Final Summary ---")
print(f"New subjects added: {new_subjects_count}")

# Calculate the updated average with the new total length of the marks list
updated_average = sum(marks) / len(marks)
print(f"Updated average marks: {updated_average:.2f}")
```

```python
# Task 3
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# Initialize variables to keep track of our summary statistics
pass_count = 0
fail_count = 0
highest_avg = 0
topper_name = ""
total_class_avg_sum = 0

# Print the formatted table header
print(f"{'Name':<18}| {'Average':<7} | {'Status'}")
print("-" * 40)

# Loop through each student in the class data
for name, marks in class_data:
    # 1. Compute their average (sum of marks divided by number of subjects)
    average = sum(marks) / len(marks)
    total_class_avg_sum += average

    # 2. Assign result status based on the average
    if average >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    # Check if current student is the new class topper
    if average > highest_avg:
        highest_avg = average
        topper_name = name

    # 3. Print the formatted row for the report
    # <18 left-aligns the name, >5.2f right-aligns the float to 2 decimal places
    print(f"{name:<18}|  {average:>5.2f}  | {status}")

# Calculate the overall class average using the sum we collected in the loop
class_average = total_class_avg_sum / len(class_data)

# Print the final summary section after the table
print("\n--- Summary ---")
print(f"Number of students who passed: {pass_count}")
print(f"Number of students who failed: {fail_count}")
print(f"Class topper: {topper_name} ({highest_avg:.2f})")
print(f"Class average: {class_average:.2f}")
```

```python
# Task 4
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# 1. Strip leading and trailing whitespace
clean_essay = essay.strip()
print("1. Stripped Essay:")
print(clean_essay)
print("-" * 40)

# 2. Convert to Title Case
print("2. Title Case:")
print(clean_essay.title())
print("-" * 40)

# 3. Count occurrences of the word "python"
python_count = clean_essay.count("python")
print("3. Count:")
print(f"The word 'python' appears {python_count} times.")
print("-" * 40)

# 4. Replace "python" with "Python 🐍"
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("4. Replaced Essay:")
print(replaced_essay)
print("-" * 40)

# 5. Split into individual sentences
sentences = clean_essay.split(". ")
print("5. Split Sentences List:")
print(sentences)
print("-" * 40)

# 6. Print each sentence on a new line, numbered, ensuring a period at the end
print("6. Numbered Sentences:")
for index, sentence in enumerate(sentences, 1):
    # Add a period if it doesn't already end with one
    if not sentence.endswith("."):
        sentence += "."
    print(f"{index}. {sentence}")
```
